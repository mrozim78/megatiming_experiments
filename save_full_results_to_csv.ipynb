```python
import pymongo
import pandas as pd
from pymongo import MongoClient
client = MongoClient()
myclient = pymongo.MongoClient("mongodb://localhost:27017/")
mydb = myclient["megatiming"]
ranking_col = mydb["result_ranking"]
athlete_col = mydb["result_athlete"]
contest_col = mydb["result_contest"]
df_ranking = pd.DataFrame(list(ranking_col.find()))
df_athlete = pd.DataFrame(list(athlete_col.find()))
df_contest = pd.DataFrame(list(contest_col.find()))
```

```python
df_result = df_ranking.merge(df_athlete,how="inner" , left_on=['uuid','resultid','eventid'], right_on = ['uuid','resultid','eventid'],copy=False)
df_result.rename(columns={"_id_x": "id_ranking", "_id_y": "id_athlete"},inplace=True)
df_result_contest = df_result.merge(df_contest,left_on=['uuid'],right_on=['uuid'],copy=False)
df_result_contest.rename(columns={"_id":"id_contest"},inplace=True)
```
