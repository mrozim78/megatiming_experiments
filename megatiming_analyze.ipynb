```python
import pymongo
import pandas as pd
from pymongo import MongoClient
client = MongoClient()
myclient = pymongo.MongoClient("mongodb://localhost:27017/")
mydb = myclient["megatiming"]
ranking_col = mydb["result_ranking"]
athlete_col = mydb["result_athlete"]
df_ranking = pd.DataFrame(list(ranking_col.find()))
df_athlete = pd.DataFrame(list(athlete_col.find()))
```

```python
df_ranking.head(10)
```

```python
df_athlete.head(10)
```

```python
df_ranking = df_ranking.merge(df_athlete,how="inner" , left_on=['uuid','resultid'], right_on = ['uuid','resultid'])
```

```python
df_ranking.head(10)
```

```python
df_age_8_free_25_male = df_ranking.query('agemax=="8" and stroke == "FREE" and swimtime!="00:00:00.00" and athlete_gender=="M" and distance=="25"')
```

```python
df_age_8_free_25_male.size
```

```python
minValue = df_age_8_free_25_male['swimtime'].min()
print('Best result in 25 m in FREE on age 8 is '+str(minValue))
```
